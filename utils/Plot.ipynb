```python
import matplotlib.pyplot as plt
from pylab import *                                 #支持中文
mpl.rcParams['font.sans-serif'] = ['SimHei']

# control cycle
x = [0.005, 0.01, 0.03, 0.05, 0.1, 0.15, 0.2]
y = [98.59, 98.52, 95.56, 94.06, 90.33, 87.62, 86.25]

#plt.plot(x, y, 'ro-')
#plt.plot(x, y1, 'bo-')
ax = plt.subplot(111)
plt.xlim(0, 0.2)  # 限定横轴的范围
plt.ylim(83, 100)  # 限定纵轴的范围
# plt.plot(x, y, marker='o', mec='r', mfc='w',label=u'y=x^2曲线图')
plt.plot(x, y, ms=10)
# plt.legend()  # 让图例生效
# 设置刻度字体大小
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
# plt.xticks(x, names, rotation=45)
# plt.margins(0)
# plt.subplots_adjust(bottom=0.15)
plt.xlabel("control cycle (s)", fontsize=14) #X轴标签
plt.ylabel("average control precision (%)", fontsize=14) #Y轴标签
plt.title("Influence of control cycle",fontsize=14) #标题

plt.show()
```
